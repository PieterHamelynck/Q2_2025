```python
import base64
import html
import re
from urllib.error import HTTPError, URLError
from urllib.parse import parse_qs, urlparse
from urllib.request import Request, urlopen

from IPython.display import HTML, display


# Extract Google Drive file id from common URL patterns.
def extract_drive_file_id(url: str) -> str | None:
    parsed = urlparse(url)
    if 'drive.google.com' not in parsed.netloc:
        return None

    if parsed.path == '/open':
        return parse_qs(parsed.query).get('id', [None])[0]

    path_parts = [p for p in parsed.path.split('/') if p]
    if len(path_parts) >= 3 and path_parts[0] == 'file' and path_parts[1] == 'd':
        return path_parts[2]

    if parsed.path == '/uc':
        return parse_qs(parsed.query).get('id', [None])[0]

    match = re.search(r'/d/([a-zA-Z0-9_-]+)', parsed.path)
    if match:
        return match.group(1)

    return None


def candidate_image_urls(url: str) -> list[str]:
    file_id = extract_drive_file_id(url)
    if not file_id:
        return [url]

    return [
        f'https://drive.google.com/thumbnail?id={file_id}&sz=w1200',
        f'https://drive.google.com/uc?export=view&id={file_id}',
        f'https://drive.google.com/uc?export=download&id={file_id}',
        url,
    ]


def fetch_image_as_data_uri(url: str) -> str | None:
    headers = {
        'User-Agent': 'Mozilla/5.0',
        'Accept': 'image/avif,image/webp,image/apng,image/*,*/*;q=0.8',
    }

    for candidate in candidate_image_urls(url):
        try:
            req = Request(candidate, headers=headers)
            with urlopen(req, timeout=12) as resp:
                content_type = (resp.headers.get('Content-Type') or '').lower()
                data = resp.read()

            # Skip HTML pages (Drive preview pages) and empty responses.
            if not data or 'text/html' in content_type:
                continue

            if not content_type.startswith('image/'):
                # Heuristic fallback for unknown content-types.
                if data[:4] == b'\x89PNG':
                    content_type = 'image/png'
                elif data[:3] == b'\xff\xd8\xff':
                    content_type = 'image/jpeg'
                elif data[:6] in (b'GIF87a', b'GIF89a'):
                    content_type = 'image/gif'
                else:
                    continue

            b64 = base64.b64encode(data).decode('ascii')
            return f'data:{content_type};base64,{b64}'

        except (HTTPError, URLError, TimeoutError, ValueError):
            continue

    return None


# Keep only the two columns we need and ignore rows with no intro text.
df_filtered = df.loc[:, ['Vertel hier wat leuks over jezelf', 'Fotootjes van jou!']].dropna(subset=['Vertel hier wat leuks over jezelf'])

cards = []
for index, row in df_filtered.iterrows():
    intro = html.escape(str(row['Vertel hier wat leuks over jezelf']).strip())
    photos_raw = row.get('Fotootjes van jou!', '')
    photo_urls = []

    if pd.notna(photos_raw):
        photo_urls = [u.strip() for u in str(photos_raw).split(',') if u.strip()]

    image_tags = []
    for i, url in enumerate(photo_urls):
        safe_url = html.escape(url, quote=True)
        data_uri = fetch_image_as_data_uri(url)

        if data_uri:
            image_tags.append(
                f'''<img src="{data_uri}" alt="Foto {i+1}" class="photo-item" loading="lazy">'''
            )
        else:
            image_tags.append(
                f'''<a href="{safe_url}" target="_blank" rel="noopener noreferrer" class="photo-fallback">'''
                f'''Foto {i+1} openen</a>'''
            )

    images_html = ''.join(image_tags)

    cards.append(f'''
        <div class="person-card">
            <h3>Persoon {index}</h3>
            <p>{intro}</p>
            <div class="photo-grid">{images_html or '<em>Geen foto beschikbaar</em>'}</div>
        </div>
    ''')

html_output = f'''
<style>
    .gallery-wrap {{
        max-width: 980px;
        margin: 18px auto;
        padding: 20px;
        border-radius: 16px;
        background: linear-gradient(180deg, #ffffff 0%, #f4f8ff 100%);
        border: 1px solid #dbe3f5;
        box-shadow: 0 10px 30px rgba(44, 62, 80, 0.08);
        font-family: "Segoe UI", Tahoma, sans-serif;
    }}
    .gallery-title {{
        margin: 0 0 14px;
        color: #1f3a63;
        font-size: 1.35rem;
    }}
    .person-card {{
        margin-bottom: 14px;
        padding: 14px;
        border-radius: 12px;
        background: #ffffff;
        border: 1px solid #e6ecfa;
    }}
    .person-card h3 {{
        margin: 0 0 8px;
        color: #224b87;
    }}
    .person-card p {{
        margin: 0 0 10px;
        color: #23374d;
        line-height: 1.5;
        white-space: pre-wrap;
        overflow-wrap: anywhere;
        word-break: break-word;
    }}
    .photo-grid {{
        display: flex;
        flex-wrap: wrap;
        gap: 10px;
    }}
    .photo-item {{
        width: 140px;
        height: 140px;
        object-fit: cover;
        border-radius: 10px;
        border: 1px solid #d3def3;
        background: #f1f5ff;
    }}
    .photo-fallback {{
        display: inline-flex;
        align-items: center;
        justify-content: center;
        width: 140px;
        height: 140px;
        border-radius: 10px;
        border: 1px dashed #8ea3cc;
        background: #f7f9ff;
        color: #224b87;
        text-decoration: none;
        font-size: 0.9rem;
        padding: 8px;
        text-align: center;
    }}
</style>
<div class="gallery-wrap">
    <h2 class="gallery-title">Fotootjes + Intro Tekst</h2>
    {''.join(cards)}
</div>
'''

display(HTML(html_output))
```
